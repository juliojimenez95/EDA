# **Fase 2: Valores Faltantes**

# Campos principales del medicamento

- **id**: identificador interno del objeto (no es de la Anvisa, pero útil en tu base de datos).  
- **codigo**: código interno de la Anvisa para el medicamento.  
- **numeroRegistro**: número de registro del medicamento en la Anvisa (ej.: `188300060`).  
- **nomeProduto**: nombre comercial del medicamento (ej.: “ácido zoledrónico”).  
- **numeroProcesso**: número del proceso administrativo en la Anvisa referente al registro.  
- **empresaNome**: nombre de la empresa responsable (titular del registro).  
- **empresaCnpj**: CNPJ de la empresa titular.  
- **principioAtivo**: sustancia farmacológicamente activa.  
- **tarja**: clasificación de la tarja (ej.: roja, negra, sin prescripción).  

---

# Presentaciones del medicamento

El campo **apresentacoes** es una lista; cada ítem describe una versión/embalaje del producto. Dentro de cada presentación:

- **codigo**: identificador de la presentación.  
- **apresentacao**: descripción comercial (ej.: “0,8 MG/ML SOL INJ IV CT FA VD TRANS X 5 ML”).  
- **formasFarmaceuticas**: forma farmacéutica (ej.: solución inyectable, comprimido).  
- **numero**: índice secuencial de la presentación.  
- **dataPublicacao**: fecha en que la presentación fue publicada en la Anvisa.  
- **validade** y **tipoValidade**: tiempo de validez del registro (ej.: 24 meses, tipo 2 = meses).  
- **registro**: número de registro específico de la presentación.  
- **principiosAtivos**: lista de principios activos usados en esa presentación.  
- **qtdUnidadeMedida**: cantidad de la sustancia (ej.: “5 mililitro”).  
- **embalagemPrimaria**: envase que contiene directamente el medicamento (puede ser `null` si está descrito en `embalagemPrimariaTodas`).  
- **embalagemSecundaria**: envase externo (ej.: cartón).  
- **embalagemPrimariaTodas**: lista detallando todos los envases primarios posibles.  
- **embalagemSecundariaTodas**: lista detallando todos los envases secundarios posibles.  
- **fabricantesNacionais** / **fabricantesInternacionais**: información de la fábrica (nombre, dirección, país, etapa de fabricación).  
- **viasAdministracao**: cómo se administra el medicamento (ej.: oral, intravenosa).  
- **ifaUnico**: indica si hay solo un IFA (insumo farmacéutico activo).  
- **conservacao**: recomendaciones de almacenamiento.  
- **restricaoPrescricao**: exigencia legal para venta (ej.: receta retenida).  
- **restricaoUso**: público objetivo (ej.: adulto, pediátrico).  
- **destinacao**: uso hospitalario, ambulatorio, etc.  
- **restricaoHospitais**: si hay restricción de venta solo para hospitales (`S/N`).  
- **medicamentoReferencia**: indica si la presentación es de referencia (`Y/N`).  
- **apresentacaoFracionada**: si permite fraccionamiento (`Y/N`).  
- **tipoAutorizacao**: estado del registro (ej.: REGISTRADO).  
- **ativa / emAnalise / inativa**: banderas que indican la situación de la presentación.  

---

# Información regulatoria y de bula

- **medicamentoReferencia**: nombre del medicamento de referencia (ej.: Zometa).  
- **classesTerapeuticas**: clase terapéutica (ej.: supresores de la reabsorción ósea).  
- **bula_txt / bula_txt_profissional**: texto bruto de la bula (puede estar vacío si no fue extraído).  
- **bula_pdf / bula_pdf_profissional**: enlaces o rutas a los PDFs de la bula para paciente y profesional.  
- **categoriaRegulatoria**: tipo de medicamento (Genérico, Similar, Biológico, etc.).  
- **situacaoRegistro**: situación del registro en la Anvisa (ej.: ACTIVO, CANCELADO).  
- **dataProduto**: fecha del primer registro o publicación del producto.  
- **dataVencimentoRegistro**: vencimiento del registro en la Anvisa (fecha de expiración).  
- **dataPublicacao**: última fecha en la que hubo publicación/actualización del registro.  
- **indicacao**: campo que podría contener la indicación terapéutica (actualmente vacío).

```python
# Explota la lista de dicts
df_exp = df_normalizado.explode('apresentacoes', ignore_index=True)

# Normaliza los dicts de 'apresentacoes' (sep='.' para aplanar niveles)
df_apr = pd.json_normalize(df_exp['apresentacoes'], sep='.')

# Detecta columnas duplicadas respecto a las originales
orig_cols = set(df_exp.columns) - {'apresentacoes'}   
dup_cols  = orig_cols.intersection(df_apr.columns)    

# Renombra solo las duplicadas con el prefijo 'apresentacoes.'
df_apr = df_apr.rename(columns={c: f"apresentacoes.{c}" for c in dup_cols})

# Concatena, conservando columnas originales y las expandidas
df_final = pd.concat([df_exp.drop(columns=['apresentacoes']), df_apr], axis=1)
```

```python
(
    df_final
    .isnull()
    .melt()
    .pipe(
        lambda df: (
            sns.displot(
                data=df,
                y="variable",
                hue="value",
                multiple="fill",
                aspect=2,
                height=10
                
        )
        )
    )
).fig.suptitle("Proporción de valores nulos y no nulos por variable", fontsize=14, y=1.02)
```

```python
cmap = ListedColormap(["black", "lightgray"])
ax = (
    df_final
    .isnull().
    transpose()
    .pipe(
        lambda df: (
            sns.heatmap(
                data=df,
                cbar=False,
                cmap=cmap,
            )
        )
    )
)
plt.title("Mapa de calor: Valores Nulos vs No Nulos", fontsize=14, pad=20)


legend_patches = [
    mpatches.Patch(color="black", label="No Nulo"),
    mpatches.Patch(color="lightgray", label="Nulo")
]

# Agregar la leyenda al gráfico
plt.legend(
    handles=legend_patches,
    loc="upper right",   
    bbox_to_anchor=(1.25, 1)  
)

plt.show()
```

Se especula que los ultimos registros del dataset sean errores o información muy incompleta que puede perjudicar el análisis.

```python
# -------------------------
# Expandir fabricantes nacionais
# -------------------------
fabricantes_nac = pd.json_normalize(df_final["fabricantesNacionais"])
fabricantes_nac = fabricantes_nac.add_prefix("fabricantesNacionais_")

# -------------------------
# Expandir fabricantes internacionais
# -------------------------
fabricantes_int = pd.json_normalize(df_final["fabricantesInternacionais"])
fabricantes_int = fabricantes_int.add_prefix("fabricantesInternacionais_")

# -------------------------
# Concatenar todo al dataframe original
# -------------------------
df_final_expandido = pd.concat(
    [df_final.drop(columns=["fabricantesNacionais", "fabricantesInternacionais"]),
     fabricantes_nac,
     fabricantes_int],
    axis=1
)

# Verificar resultado
print(df_final_expandido.head())

```
