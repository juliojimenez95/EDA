```python
import numpy as np
import matplotlib.pyplot as plt

# Reproducibilidad
rng = np.random.default_rng(42)

# 1. Rango del eje X (salida del modelo)
z = np.linspace(-20, 20, 400)
p = 1 / (1 + np.exp(-z / 4))  # funci√≥n sigmoide

# 2. Puntos aleatorios
n_points = 500
z_rand = rng.uniform(0, 20, n_points)

# L√≠nea diagonal que separa clases: de (0,0) a (20,1)
diagonal = z_rand / 20

# 3. Crear clases:
#    - 1's por encima de la diagonal
#    - 0's por debajo
y = rng.random(n_points)
labels = np.where(y > diagonal, 1, 0)

# 4. Agregar dispersi√≥n vertical para mejor visualizaci√≥n
y_values = np.where(
    labels == 1,
    y + rng.uniform(0.02, 0.05, n_points),  # un poco por encima
    y - rng.uniform(0.02, 0.05, n_points)   # un poco por debajo
)
y_values = np.clip(y_values, 0, 1)  # limitar entre 0 y 1

# 5. Graficar
#plt.figure(figsize=(7,4))
plt.scatter(z_rand[labels==0], y_values[labels==0], s=20,  label="0's", alpha=0.8)
plt.scatter(z_rand[labels==1], y_values[labels==1], s=20,  label="1's", alpha=0.8)
#plt.plot(z, p, color='crimson', linewidth=2.5, label="Sigmoid curve")

# Dibujar la diagonal separadora
plt.plot([0, 20], [0, 1], '--', color='gray', linewidth=2, label='Limite')

plt.xlim(0, 20.1)
plt.ylim(0, 1.01)
plt.legend(loc= 'upper left')
for spine in ['top','right']:
    plt.gca().spines[spine].set_visible(False)
plt.tight_layout()
plt.show()
```
